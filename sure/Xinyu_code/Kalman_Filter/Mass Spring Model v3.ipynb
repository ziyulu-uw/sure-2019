### The SDE of the Model

<ul>
    <li> The dynamic system composed of a spring and a object with fraction and is a 2D Ornstein-Uhlenbeck process
        $$\begin{pmatrix}\dot{x}\\\dot{v}\end{pmatrix}= \begin{pmatrix}0 & 1 \\ -\frac{k}{m}  & -\frac{\gamma}{m}
        \end{pmatrix}\begin{pmatrix}x\\v\end{pmatrix} dt+ \begin{pmatrix}0 & 0 \\0 &
        \sigma\end{pmatrix}\begin{pmatrix}dW_t^{(1)}\\dW_t^{(2)}\end{pmatrix}$$
   </li>
   
   <li> Let $X(t)= \begin{pmatrix}\dot{x}\\\dot{v}\end{pmatrix}, \alpha = \begin{pmatrix}0 & 1 
    \\ -\frac{k}{m}  & -\frac{\gamma}{m}\end{pmatrix}$, $B = \begin{pmatrix}0 & 0 \\0 & \sigma\end{pmatrix}$,
    $dW_t =  \begin{pmatrix}dW_t^{(1)}\\dW_t^{(2)}\end{pmatrix}$, then we have
     $$dX = \alpha X dt + BdW_t$$
   </li>
   
   <li> The solution to the SDE is $$X_t = e^{\alpha t}X_0 + \int_0^te^{(t-s)\alpha}BdW_s$$
    Let $A(t) = e^{\alpha t}$, then $X_t = A(t)X_0 + \int_0^tA({t-s})BdW_s$
<uli>


        
### Spectral Decomposition on Fundamental Solution A(t)
       
<ul> 
    <li> $A(t) = UDU^{-1}$, where the columns of $U$ are the right eigenvectors of $S$. $D$ is a diagonal matrix with eigenvalues $\lambda_1$ and $\lambda_2$ on the diagonal.</li>
    <li> $A(t)=e^{\alpha t} =\sum \frac{t^{n}}{n !} \alpha^{n}
    =\sum \frac{t^{n}}{n !} U D^{n} U^{-1} 
    =U\left(\sum \frac{t^{n}}{n !} D^{n}\right) U^{-1}
    =U\left(\begin{array}{cc}{e^{\lambda_{1}t} } & {0} \\ {0} & {e^{\lambda_{2}t}}\end{array}\right) U^{-1}$</li>
    <li> $\lambda_1 = -\frac{\gamma}{2m} + i\sqrt{\frac{4km-\gamma^2}{4m^2}} =-\mu + i\omega$,
        $\lambda_2 = -\frac{\gamma}{2m} - i\sqrt{\frac{4km-\gamma^2}{4m^2}} =-\mu - i\omega$,
        $U = \begin{pmatrix} 1 & 1 \\ \lambda_1 & \lambda_2 \end{pmatrix}$, thus
        $\begin{align*}
        &A(t) = U\left(\begin{array}{cc}{e^{\lambda_{1} t}} & {0} \\ {0} & {e^{\lambda_{2}t}}\end{array}\right)U^{-1} \\
        &= \left(\begin{array}{ll}{1} & {1} \\ {\lambda_{1}} & {\lambda_{2}}\end{array}\right)\left(\begin{array}{cc}{e^{\lambda_{1} t}} & {0} \\ {0} &{e^{\lambda_{2}t}}\end{array}\right)\left(\lambda_{2}-\lambda_{1}\right)^{-1}\left(\begin{array}{cc}{\lambda_{2}} & {-1} \\ {-\lambda_{1}} & {1}\end{array}\right) \\
        &= \frac{1}{\lambda_{2}-\lambda_{1}}\left(\begin{array}{ll}{\lambda_{2} e^{\lambda_{1} t}-\lambda_{1} e^{\lambda_{2} t}} & {-e^{\lambda_{1} t}+e^{\lambda_{2} t}} \\ {\lambda_{1} \lambda_{2}\left(e^{\lambda_{1} t}-e^{\lambda_{2} t}\right)} & {-\lambda_{1} e^{\lambda_{1} t}+\lambda_{2} e^{\lambda_{2} t}}\end{array}\right) \end{align*}$</li>
        

### Covariance Matrix of the Noise

<ul> 
    <li> Suppose $X_{n} = X_t$, $X_{n+1} = X_{t+\Delta t}$, then 
        $$X_{n+1}-e^{\alpha \Delta t} X_{n}=\int_{0}^{t+\Delta t} e^{(t+\Delta t-s) \alpha} B dW_{s}-\int_{0}^{t} e^{(t+\Delta t-s) \alpha} BdW_s =\int_{t}^{t+\Delta t} e^{(t+\Delta t-s) \alpha} B d W_{s}.$$ So 
        $$X_{n+1}= e^{\alpha \Delta t} X_{n} + \int_{t}^{t+\Delta t} e^{(t+\Delta t-s) \alpha} B d W_{s} = A(\Delta t) X_n + \int_t^{t+\Delta t}A(t+\Delta t -s)BdW_s. $$ </li>
    <li> Let $W_n = \int_t^{t+\Delta t}A(t+\Delta t -s)BdW_s$, $A(t+\Delta t -s) = \begin{pmatrix} a_{11} & a_{12} \\ a_{21} & a_{22}\end{pmatrix}$. Then, 
    $$W_n = \begin{pmatrix} \int_t^{t+\Delta t} \sigma a_{12} dW_s^{(2)} \\ \int_t^{t+\Delta t} \sigma a_{22} dW_s^{(2)}\end{pmatrix}. $$ Thus, the covariance matrix $R_n$ of $W_n$ is
    $$R_n =\mathbb{E}[W_nW_n^T] = \begin{pmatrix}\int_t^{t+\Delta t} \sigma^2 a_{12}^2 ds & \int_t^{t+\Delta t} \sigma^2 a_{12}a_{22} ds \\ \int_t^{t+\Delta t} \sigma^2 a_{12}a_{22}  ds & \int_t^{t+\Delta t} \sigma^2 a_{22}^2 ds\end{pmatrix} 
    =  \left(\frac{\sigma}{\lambda_{2}-\lambda_{1}}\right)^{2}
    \begin{pmatrix}
    \frac{e^{2 \lambda_{1} \Delta t}-1}{2 \lambda_{1}}+\frac{e^{2 \lambda_{2} \Delta t}-1}{2 \lambda_{2}}-2 \frac{e^{\left(\lambda_{1}+\lambda_{2}\right) \Delta t}-1}{\left(\lambda_{1}+\lambda_{2}\right)} 
    & \frac{e^{2 \lambda_{1} \Delta t}-1}{2 }+\frac{e^{2 \lambda_{2} \Delta t}-1}{2 }- ({e^{\left(\lambda_{1}+\lambda_{2}\right) \Delta t}-1})  \\
    \frac{e^{2 \lambda_{1} \Delta t}-1}{2 }+\frac{e^{2 \lambda_{2} \Delta t}-1}{2 }- ({e^{\left(\lambda_{1}+\lambda_{2}\right) \Delta t}-1})  & 
     \lambda_{1}\frac{e^{2 \lambda_{1} \Delta t}-1}{2}+\lambda_{2}\frac{e^{2 \lambda_{2} \Delta t}-1}{2 }-2 \lambda_{1}\lambda_{2}\frac{e^{\left(\lambda_{1}+\lambda_{2}\right) \Delta t}-1}{\left(\lambda_{1}+\lambda_{2}\right)} \end{pmatrix}$$
     Here we can see $R_n$ does not depend on time. Let denote $R=R_n$, $n=1,2,3...$

##### Remark:
<li> In professor's note "Linear gaussian dynamics and filtering", Linear noisy dynamics $X_{n+1} =AX_n +W_n$, $A$ refers to $A(\Delta t)$ here, which is a constant matrix once $\Delta t$ is fixed.</li>
<li> $W_n$ is a gaussian noise with mean $0$ and constant covariance matrix $R$ 

```python
import numpy as np
from numpy import linalg as la
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.integrate import simps
```

```python
def transpose(x):
    '''a function used to take transpose of each matrix inside a tensor x
     x is a tensor n*d*c, return n*c*d'''
    d = len(x[0,:,0])                       #num of rows of each matrix
    c = len(x[0,0,:])                       #num of columns of each matrix
    n = len(x[:,0,0])                       #num of matrices in the tensor x
    
    X_T = np.zeros([n,c,d],dtype = complex)  #allocate space
    for i in range(d):
        X_T[:,:,i]=x[:,i,:]
    return X_T
```

```python
def mass_spring_process(x0,v0,n,N,k,m,gamma,sigma,h):
    '''simulate n independent paths, each with step N. 
    The state evolution satisfies Xn+1 = AXn + Wn'''
    
    ##allocate space
    X     = np.zeros([n,2,1])                    #each matrix is 2*1, has n matrices in X
    A     = np.zeros([n,2,2],dtype = complex)    #fundamental solution n*2*2
    alpha = np.zeros([2,2])                      #the alpha matrix in SDE(see in description above)
    R     = np.zeros([2,2],  dtype = complex)    #covariance matrix for W
    X_mat = np.zeros([n,N,2],dtype = complex)    #a matrix to record all the trajectory of X=[xn,vn]
    
    ##initialize X
    X[:,0,0] = x0 #each matrix is [x0;
    X[:,1,0] = v0 #                v0] 
    
    ##initialize alpha
    alpha[0,1] = 1
    alpha[1,0] = -k/m
    alpha[1,1] = -gamma/m
    
    ##initialize A = exp(alpha*dt)
    w, U = la.eig(alpha)                      #spectral decomposition: dot(a[:,:], U[:,i]) = w[i] * U[:,i] 
                                              #w are eigenvalues, the columns of U are eigenvectors
    D = np.diag(np.exp(w*h))                  #take the exponential of eigenvalue*dt and put them on the diagonal
    A_= U@D@la.inv(U)                         #the fundamental solution A(dt)
    A[:,0,0] = A_[0,0]                        #convert A_ from a 2*2 matrix to n*2*2 tensor, 
    A[:,0,1] = A_[0,1]                        #each matrix in tensor is the same
    A[:,1,0] = A_[1,0]
    A[:,1,1] = A_[1,1]
    
    ##initialize covariance matrix R of W
    w1 = w[0]                                 #eigenvalue
    w2 = w[1]
    e1 = np.exp(2*w1*h) - 1                   #temp1
    e2 = np.exp(2*w2*h) - 1                   #temp2
    e3 = np.exp((w1+w2)*h) - 1
    R[0,0] = e1/(2*w1)+e2/(2*w2)-2*e3/(w1+w2) #assign values
    R[1,0] = e1/2+e2/2-e3    
    R[0,1] = R[1,0]
    R[1,1] = w1*e1/2+w2*e2/2-2*w1*w2*e3/(w1+w2)
    R = R*(sigma/(w2-w1))**2
    
    for i in range(N): #time evolution
        ##generate noise W
        W = np.random.multivariate_normal(np.array([0,0]),np.real(R),n) #mean is [0,0] for Wn
        W = np.reshape(W,[n,2,1])                                       #reshape W to make it n*2*1
        
        ##time evolution with noise
        X = A@X+W
        X_mat[:,i,0] = X[:,0,0].T
        X_mat[:,i,1] = X[:,1,0].T

    #print(np.size(X_mat))
    return A,R,np.real(X_mat) #the imaginary part of X are almost zero (in theory should be zero)
```

### Fixed Point Exercise
(from Professor Goodman's note "Linear gaussian dynamics and filtering")

${\bf Theroem.}$ Suppose $\mathbb{E}[X_n]=0$ for convenience. If the linear dynamics are strongly stable and $S = \mathbb{E}[X_nX_n^T]$ then
there is an $S$ with $S_n \to S$ as $n \to \infty$.
$S$ is a fixed point of the covariance dynamics.
The PDF of $X_n$ has a limit as $n \to \infty$, which is Gaussian with mean zero and covariance $S$.

${\bf Proof.}$ (sketch) The covariance satisfies 
$S_n = \sum_{k=0}^{n-1} A^k R \left(A^k\right)^t + A^nS_0 \left(A^n\right)^t \; .$
       
If the linear dynamics are exponentially stable, this sum converges to
$S = \sum_{k=0}^{\infty} A^k R \left(A^k\right)^t \; .$
In that case, the following calculation is valid:
$\begin{align*}
     \sum_{k=0}^{\infty} A^k R \left(A^k\right)^t  &= \sum_{k=1}^{\infty} A^k R \left(A^k\right)^t + R \\
     &= A \left[ \sum_{k=0}^{\infty} A^k R \left(A^k\right)^t \right] A^t + R \\
     &= ASA^t + R \; .
\end{align*}
$

${\bf Exercise.}$ Verify this by simulation in the damped oscillator example, which has $d=2$.


```python
##initialization
x0    = 2                     #initial position
v0    = 0                     #initial velocity

n     = int(5e+0)             #num of paths
t1    = 9                     #end time
h     = 0.01                  #time interval delta t
N     = int(t1/h+1)           #the num of steps of each path
t     = np.linspace(0, t1, N) #time steps (0, t1, t1/h+1)

k     = 1                   #spring constant
m     = 1                     #mass
gamma = 0.3                  #fraction coefficient
sigma = 0.1                   #noise coefficent in SDE

##simulate the process
A,R,X_mat = mass_spring_process(x0,v0,n,N,k,m,gamma,sigma,h)
#A fundamental solution   n*2*2
#R covariance matrix of W 2*2
#X_mat trajectory of n paths [xn,vn] (n*N*2)

X     = np.reshape(X_mat[:,-1,:],[n,2,1])         #Xn of n paths at t1, each of the form [xn;vn]
X_T   = transpose(X)

##calculate the covariance matrix S based on simulation result
EX    = np.average(X,axis = 0)        #get the expectation of X
S     = (X-EX)@(X_T-EX.T)     #get the matrices
ES    = np.average(S,axis=0)  #take the average and get the covariance

##calculate the covariance matrix S based on fixed point theorem
A_T   = transpose(A)
EASA  = np.average(A@S@A_T,axis=0)

##compare the result
print("The covariance matrix S based on simulation result: \n",ES)
print()
print("The covariance matrix S based on fixed point theroem: \n",EASA+R)
```

### Kalman Filter Application

Dynamics: 

  $$X_{n+1} = AX_n + W_n $$

Observation (measurement):

  $$ Z_{n+1} = C X_{n+1} + V_{n+1} \; .$$

Noise model:
<li> $W_n$ and $V_n$ independent and Gaussian </li>
<li> $\mathbb{E}[W_n] =0$  </li>
<li> $\mathbb{E}[V_n] =0$  </li>
<li> $\mathbb{E}[W_nW_n^T] =R$  </li>
<li> $\mathbb{E}[V_nV_n^T] =Q$  </li>

Let $Y_n$ be the prediction error: $Y_n = \hat{X}_n-X_n$.
Let $T_n = \text{E}\!\left[Y_nY_n^t\right] =$  covariance matrix of prediction error. 

We want to find $K_n$ to minimize $\mathbb{E}[Y_nY_n^T]$, which is equivalent to make $Y_n$ independent of $Z_0,Z_1,...,Z_n$

Therefore,
$$0 = \mathbb{E}\left[(Y_{n+1})(Z_{n+1})^T\right]=
\mathbb{E}\left[\left(X_{n+1}-\hat{X}_{n+1}\right)\left(C X_{n+1}+V_{n+1}\right)^{\top}\right]
.$$

After some calculation, we can find that
$T_{n+1} = \left( A - K_nCA\right) T_n \left( A - K_nCA\right)^t + \left( I - K_nC\right) R \left( I - K_nC\right)^t + K_nQK_n^t \; .$

$$K_n = \left( AT_nA^t + R\right)C^t \left( CAT_nA^tC^t + CTC^t + Q\right)^{-1} \; .$$


```python
def measurement(X_mat,Q):
    '''get noisy measurements'''
    # @return: noisy measurements (only observe position). Z is n*N
    # @parameter: sol is the true trajectory with evolutioin noise, n*N
    #             calculated by simulating with fundamental solution
   
    noise = np.random.normal(0,np.sqrt(Q),[n,N])    #generate independent gaussian noise
    Z = X_mat+noise                                #return the noisy measurements
    return Z
```

```python
def Mass_Spring_ODE(y, t, omega, mu):
    '''the function is used in python's package odeint
       to solve the ode'''
    x, v = y                                        #Let y be the vector [x, v]
    dydt = [v,-mu*v-omega*x]
    return dydt
```

```python
def KalmanFilter(x0,v0,A,Z,R,Q,n,N):
    '''the function to do Kalman Filter 
    @parameter: x0, v0 are initial conditions
                A fundamental solution
                Z measurements
                R covariance matrix of process noise
                Q covariance matrix of observation noise
                n num of paths
                N length of each path
    @return:    pos n*N*1 all the estimates of position
                K_val n*N*2 all the Kalman Gain
                T_val n*N*3 all the Covariance of prediction error'''
    
    ## allocate space for arrays
    Zn_hat    = np.zeros([n,2,1])    # predicted observation using old data Z(n+1) = CA X_hat(n)
    Xn_hat    = np.zeros([n,2,1])    # a posteri estimate of X
    Tn        = np.zeros([n,2,2])    # covariance matrix of prediction error
    Kn        = np.zeros([n,2,1])    # Kalman Gain
    
    pos       = np.zeros([n,N,1])    # a tensor to take all position values
    K_val     = np.zeros([n,N,2])    # a tensor to take all K11 and K12 values
    T_val     = np.zeros([n,N,3])    # a tensor to take all T11, T12, T22 values (T12 = T21)
    X_val     = np.zeros([n,N,2])    # a tensor to take all Xn = [xn;vn]
    
    ## initial state, each matrix has the form: [x0; v0]
    Xn_hat[:,0,0] = x0                   # put the initial condition into Xn_hat
    Xn_hat[:,1,0] = v0 
    
    ## create n constant matrix [1,0]
    C         = np.zeros([n,1,2])
    C[:,0,0]  = 1
    C_T       = transpose(C)
    A_T       = transpose(A)          # fundamental solution
    
    ## create n identity matrix I=[[1,0];[0,1]]
    I         = np.zeros([n,2,2])
    I[:,0,0]  = 1
    I[:,1,1]  = 1

    ## time update
    for k in range(N):
        '''The formulas are from Professor Goodman's notes Kalman Filter Formula'''
        
        ## KALMAN FILTER UPDATE
        #predicted observation using old date
        Zn_hat      = C@A@Xn_hat    
        #update covariance matrix of Yn
        Tn          = (A-Kn@C@A)@Tn@transpose(A-Kn@C@A)+(I-Kn@C)@R@transpose(I-Kn@C)+Kn@Q@transpose(Kn) 
        #update Kalman Gain Kn
        Kn          = (A@Tn@A_T+R)@C_T@la.inv(C@A@Tn@A_T@C_T+C@R@C_T+Q)\
        #update prediction using dynamics and new measurement
        z           = np.reshape(Z[:,k],[n,1,1])   #(convert measurements Z to n*1*1)
        Xn_hat      = A@Xn_hat+Kn@(z-Zn_hat)
       
        ## DATA MODIFICATION
        #modify Xn Kn Tn to be real
        Xn_hat      = np.real(Xn_hat)
        Kn          = np.real(Kn)
        Tn          = np.real(Tn)
        
        ## RECORD STATISTICS
        #record the estimation of n paths
        #pos[:,k,0]  = Xn_hat[:,0,0].T              #pos[i,:,:] is one trajectory, 
                                                   #pos[:,k,:] is all pos of n paths at time k
        #record the values of Kn of n paths
        K_val[:,k,0]= Kn[:,0,0].T                  #K11
        K_val[:,k,1]= Kn[:,1,0].T                  #K12
        
        #record the values of Tn of n paths
        T_val[:,k,0]= Tn[:,0,0].T                  #T11
        T_val[:,k,1]= Tn[:,1,0].T                  #T12 = T21
        T_val[:,k,2]= Tn[:,1,1].T                  #T22
        
        #record Xn_hat = [xn;vn] of n paths
        X_val[:,k,0]= Xn_hat[:,0,0].T              #xn
        X_val[:,k,1]= Xn_hat[:,1,0].T              #vn

    return K_val,T_val,X_val
```

#### <li> Verify that Kalman Gain $K_n$, Covariance Matrix $T_n$ stays constant in the steady state

```python
##initialization
x0    = 1                     # initial position
v0    = 0                   # initial velocity
n     = int(1e+3)             # num of paths
t1    = 40                    # end time
h     = 0.5                  # time interval delta t
N     = int(t1/h+1)           # the num of steps of each path
t     = np.linspace(0, t1, N) # time steps (0, t1, t1/h+1)
k     = 0.5                   # spring constant (mass=1, k=omega)
m     = 1                     # mass
gamma = 0.1                  # fraction coefficient (mass=1, gamma=mu)
sigma = 1                   # noise coefficent in SDE
Q     = np.ones([1,1])*0.5   # Var_v: measurement noise variance

##simulate the process
A,R,X_mat = mass_spring_process(x0,v0,n,N,k,m,gamma,sigma,h)
#A      fundamental solution 2*2;
#R      covariance matrix of W 2*2; 
#X_mat  trajectory of n paths (only position) n*N.

##get noisy measurements Z n*N
Z     = measurement(X_mat[:,:,0],Q) #measurement function only take positions as parameters

##get the theoretical solution to the ODE use Python Package
omega = k/m                   #  w = k/m, spring constant/mass
mu    = gamma/m               #  mu=gamma/m, fraction coefficient/mass
y0    = [x0,v0]               #  [initial pos, initial velocity]
sol   = odeint(Mass_Spring_ODE, y0, t, args=(omega, mu)) #theoretical solution to ODE

##KALMAN FILTER
K_val,T_val,X_val = KalmanFilter(x0,v0,A,Z,R,Q,n,N) 
#K_val n*N*2, T_val n*N*3, X_val n*N*2
pos = X_val[:,:,0]            # n*N

## STATISTICS
Epos   = np.average(pos,axis = 0)              # 1*N
EK_val = np.average(K_val,axis = 0)            # N*2
VarK1  = np.var(K_val[:,:,0],axis = 0)         # the variance of each column, 
                                               # which gives the variance of n paths at each time
VarK2  = np.var(K_val[:,:,1],axis = 0)         # the variance of each column, 
                                               # which gives the variance of n paths at each time
ET_val = np.average(T_val,axis = 0)            # N*3
VarT1  = np.var(T_val[:,:,0],axis = 0)         # the variance of each column, 
                                               # which gives the variance of n paths at each time
VarT2  = np.var(T_val[:,:,1],axis = 0)         # the variance of each column, 
                                               # which gives the variance of n paths at each time
VarT3  = np.var(T_val[:,:,2],axis = 0)         # the variance of each column, 
                                               # which gives the variance of n paths at each time

## Estimation Result Plot
plot_start = 0                               # ignore the transition part
plt.plot(t[plot_start:],Epos[plot_start:],'-',label="Kalman Filter Estimation")
plt.plot(t[plot_start:],sol[:,0][plot_start:],'--',label="Theoretical Trajectory")
plt.title("Kalman Filter Estimation")
plt.legend()
plt.show()

## Verify that Kalman Gain stays constant in the steady state
plot_start = 0      # include the transition part
plt.plot(t[plot_start:],EK_val[plot_start:,0],'-',label = "K11: avg="+str(np.around(np.average(EK_val[plot_start:,0]),decimals = 3)))
plt.plot(t[plot_start:],EK_val[plot_start:,1],'-',label = "K12: avg="+str(np.around(np.average(EK_val[plot_start:,1]),decimals = 3)))
plt.title("Expectation of the Values of Kalman Gain")
plt.xlabel("time")
plt.ylabel("E[K]")
plt.legend()
plt.show()

print("The maximum variance of K11 from 0~t1:",np.max(VarK1[plot_start:]))
print("The maximum variance of K12 from 0~t1:",np.max(VarK2[plot_start:]))

## Plot the entries of Tn as a function of time
plot_start = 0      # include the transition part
plt.plot(t[plot_start:],ET_val[plot_start:,0],'-',label = "T11: avg="+str(np.around(np.average(ET_val[plot_start:,0]),decimals = 3)))
plt.plot(t[plot_start:],ET_val[plot_start:,1],'-',label = "T12: avg="+str(np.around(np.average(ET_val[plot_start:,1]),decimals = 3)))
plt.plot(t[plot_start:],ET_val[plot_start:,2],'-',label = "T22: avg="+str(np.around(np.average(ET_val[plot_start:,2]),decimals = 3)))

plt.title("Expectation of the Values of Tn as a function of time")
plt.xlabel("time")
plt.ylabel("E[T]")
plt.legend()
plt.show()

print("The maximum variance of T11 from 0~t1:",np.max(VarK1[plot_start:]))
print("The maximum variance of T12 from 0~t1:",np.max(VarK2[plot_start:]))
```

###### Findings:

<li> In the steady state, the value of $K_n$ and $T_n$ does not change as the time goes </li>

#### <li> Verify the Fundamental Solution is correct by comparing it to result from Python ODEint package  

```python
def mass_spring_process_no_noise(x0,v0,n,N,k,m,gamma,h):
    '''simulate n independent paths, each with step N. 
    The state evolution satisfies Xn+1 = AXn'''
    
    ##allocate space
    X     = np.zeros([n,2,1])                 #each matrix is 2*1, has n matrices in X
    A     = np.zeros([n,2,2])  #fundamental solution n*2*2
    alpha = np.zeros([2,2])                   #the alpha matrix in SDE(see in description above)
    X_mat = np.zeros([n,N])    #a matrix to record all the trajectory of position x
    
    ##initialize X
    X[:,0,0] = x0 #each matrix is [x0;
    X[:,1,0] = v0 #                v0] 
    X_mat[:,0] = X[:,0,0].T
    
    #print(x0,X)
    ##initialize alpha
    alpha[0,1] = 1
    alpha[1,0] = -k/m
    alpha[1,1] = -gamma/m
    
    ##initialize A = exp(alpha*dt)
    w, U = la.eig(alpha)                      #spectral decomposition: dot(a[:,:], U[:,i]) = w[i] * U[:,i] 
                                              #w are eigenvalues, the columns of U are eigenvectors
    D = np.diag(np.exp(w*h))                  #take the exponential of eigenvalue*dt and put them on the diagonal
    A_= np.real(U@D@la.inv(U))                       #the fundamental solution A(dt)
    A[:,0,0] = A_[0,0]                        #convert A_ from a 2*2 matrix to n*2*2 tensor, 
    A[:,0,1] = A_[0,1]                        #each matrix in tensor is the same
    A[:,1,0] = A_[1,0]
    A[:,1,1] = A_[1,1]
   
    for i in range(N-1): #time evolution  
        ##time evolution with noise
        i=i+1
        X = A@X
        X_mat[:,i] = X[:,0,0].T
    #print(X_mat)
    return A,X,X_mat #the imaginary part of X are almost zero (in theory should be zero)
```

```python
##initialization
x0    = 1                     #initial position
v0    = 0                     #initial velocity

n     = int(1e+0)             #num of paths
t1    = 79.0                    #end time
h     = 0.05                  #time interval delta t
N     = int(t1/h+1)           #the num of steps of each path
t     = np.linspace(0, t1, N) #time steps (0, t1, t1/h+1)

k     = 0.5                   #spring constant (mass=1, k=omega)
m     = 1                     #mass
gamma = 0.01                  #fraction coefficient (mass=1, gamma=mu)
sigma = 0.1                   #noise coefficent in SDE

Q     = np.ones([1,1])*0.03   #Var_v: measurement noise variance

##simulate the process
A,X,X_mat = mass_spring_process_no_noise(x0,v0,n,N,k,m,gamma,h)
#A      fundamental solution 2*2;
#R      covariance matrix of W 2*2; 
#X      simulation result at t1 n*2*1;
#X_mat  trajectory of n paths (only position) n*N.
X_mat = np.real(X_mat)

##get the theoretical solution to the ODE use Python Package
omega = k/m                   #  w = k/m, spring constant/mass
mu    = gamma/m               #  mu=gamma/m, fraction coefficient/mass
y0    = [x0,v0]               #  [initial pos, initial velocity]
sol   = odeint(Mass_Spring_ODE, y0, t, args=(omega, mu)) #theoretical solution to ODE

plot_start = 0      #ignore the transition part
plt.plot(t[plot_start:],X_mat[0,plot_start:],'-',label="Fundamental Solution")
plt.plot(t[plot_start:],sol[:,0][plot_start:],'--',label="Python ODE Package")
plt.title("Comparison Between Fundamental Solution Evolution and Python ODE Package")
plt.legend()
plt.show()

plt.title("The absolute difference between python package and fundamental solution simulation")
plt.plot(t[plot_start:],X_mat[0,plot_start:]-sol[:,0][plot_start:],'-',label="Fundamental Solution")
plt.show()
```

#### <li> Verify the covariance matrix of $W_n$ is correct by comparing it with Python integration package

```python
## constants
omega   = 1.5                               # omega = k/m
mu      = .2                                # mu  = gamma/m
w       = np.sqrt(omega - 0.25*mu**2)
N       = 200                               # number of time steps in one simulation
dt      = .01                               # step size in one simulation
sigma   = .1                                # noise coefficient in SDE

## eigenvalues of A
lambda1 = complex(-0.5*mu,w)
lambda2 = complex(-0.5*mu,-w) 

s       = np.linspace(0,dt,100)             # dummy variable s inside the integration
dt_s    = dt-s

# get the value of A12,A22 for integration
A12     = -np.exp(lambda1*dt_s) + np.exp(lambda2*dt_s)
A22     = -lambda1*np.exp(lambda1*dt_s) + lambda2*np.exp(lambda2*dt_s)
A12     = A12/(lambda2-lambda1)
A22     = A22/(lambda2-lambda1)

## use Numpy Simpson Integration to calculate entries of R
Rs11    = sigma**2*simps(A12**2,s)
Rs12    = sigma**2*simps(A12*A22,s)
Rs22    = sigma**2*simps(A22**2,s)

## construct covariance matrix R using the formula we derive
e1      = np.exp(2*lambda1*dt)
e2      = np.exp(2*lambda2*dt)
e3      = np.exp((lambda1 + lambda2)*dt)
R11     = 0.5*(e1-1)/lambda1 + 0.5*(e2-1)/lambda2 - 2*(e3-1)/(lambda1 + lambda2)
R12     = 0.5*(e1-1) + 0.5*(e2-1) - (e3-1)
R21     = R12
R22     = lambda1*0.5*(e1-1) + lambda2*0.5*(e2-1) - 2*lambda1*lambda2*(e3-1)/(lambda1 + lambda2)
R11     = R11*(sigma/(lambda2-lambda1))**2
R12     = R12*(sigma/(lambda2-lambda1))**2
R22     = R22*(sigma/(lambda2-lambda1))**2

## show the difference
print("The entries in R using \n \t Simpson Integration and Formula:",)
print("R11:",np.real(Rs11),'\t',np.real(R11))
print("R12:",np.real(Rs12),'\t',np.real(R12))
print("R22:",np.real(Rs22),'\t',np.real(R22))
```
